Conditional structures

```python
# Exercise 3: Conditionals
# Given the variable `temperature`, write an if-else statement that:
# - Prints "It's hot!" if temperature is greater than 30
# - Prints "It's warm!" if temperature is between 20 and 30 (inclusive)
# - Prints "It's cold!" otherwise

temperature = 5  # Try changing this value

# You should use if, elif and else.
# For printing use print()\\

if temperature > 30:
  print('Its hot')
elif 20 <= temperature <= 30:
  print('Its warm')
else:
  print('Its cold')

# Exercise 4: Odd or Even
# Write a condition to check if `number` is even or odd
# If it's even, print "Even number"
# Otherwise, print "Odd number"

number = 16  # Try changing this value

if number % 2 == 0 :
    print("Even number")
else:
    print("Odd number")
```

```python
# Exercise 7: Lists
# Define a list `fruits` with elements: "apple", "banana", "cherry"
# Append "orange" to the list
# Print the second item in the list

fruits = ["apple","banana","cherry"]
fruits.append("orange") # Add "orange"
print(fruits[1])  # Print second item

#BonusTrack by Beib

fruits = ["apple","banana","cherry"]
fruits.insert(1, "orange") # Add "orange as a second item"
print(fruits[1])  # Print second item

# Exercise 8: Dictionaries
# Define a dictionary `student` with keys "name" (value: "Alice"), "age" (value: 22)
# Add a key "grade" with value "A"
# Print the age of the student

student = {"name": "Alice", "age": 22}
student["grade"] = "Senior"  # Add "grade"
print(student["age"])  # Print age
print(student["grade"])

# Exercise 9: Sets
# Define a set `numbers` with values {1, 2, 3, 4}
# Add the number 5 to the set
# Try adding 3 again and print the set

numbers = {1, 2, 3, 4}
numbers.add(5) # Add 5
numbers.remove(3)
print(numbers)
numbers.add(3) # Add 3 again
print(numbers)

# Exercise 10: Tuples
# Define a tuple `colors` with values ("red", "green", "blue")
# Print the last color in the tuple

colors = ("red", "green", "blue")
print(colors[2])  # Print last color
```
